# Data Preprocessing

### - Handling Missing Values


```python
import pandas as pd
import re

rawBooks_df = pd.read_csv("raw_books.CSV")

# pattren for n/a 
pattern = '\u200en\u200e/\u200ea\u200e'

# Search for this pattren
specific_hidden_char_rows = rawBooks_df[rawBooks_df.apply(lambda x: x.astype(str).str.contains(pattern, na=False)).any(axis=1)]

# display
specific_hidden_char_rows
```

```python

# replace this pattren with 'Unknown'
for col in rawBooks_df.columns:
    rawBooks_df[col] = rawBooks_df[col].apply(
        lambda x: 'Unknown' if isinstance(x, str) and pattern in x else x
    )
```

```python
# Check
unkown_author = rawBooks_df[rawBooks_df.apply(lambda x: x.astype(str).str.contains('Unknown', na=False)).any(axis=1)]
unkown_author
```

```python
missing_rows = rawBooks_df[(rawBooks_df == 'Null').any(axis=1) | 
                        (rawBooks_df == 'Unknown').any(axis=1) | 
                        (rawBooks_df.isna()).any(axis=1)]

#  Count the number of "Null", "Unknown", and NaN values in each column
missing_counts = ((rawBooks_df == 'Null') | (rawBooks_df == 'Unknown')).sum() + rawBooks_df.isna().sum()

#  Display the count of missing values in each column
print("\nCount of missing values ('Null', 'Unknown', or NaN) in each column:")
missing_counts
```

```python
print("Rows with 'Null', 'Unknown', or empty (NaN) values:")
missing_rows
```

```python
rawBooks_df['Rating'] = rawBooks_df['Rating'].replace('Null', float('nan')).fillna(0)
rawBooks_df['Num Of Reviews'] = rawBooks_df['Num Of Reviews'].replace('Null', float('nan')).fillna(0)
rawBooks_df = rawBooks_df.dropna(subset=['Author'])

rawBooks_df[['Rating', 'Num Of Reviews','Author']] 

# NOTE: Need to rearrrange the index after the drop 
# NOTE : Need to explain the 3 cases of Missing value 
# NOTE : Need a code of pattren n/a ?
```

### - Format for Price , Rating , Num Of Reviews 

```python
# Cleaning the columns
rawBooks_df['Price'] = rawBooks_df['Price'].astype(str).str.extract(r'(\d+\.?\d*)')  # Extracting numeric part from Price
rawBooks_df['Rating'] = rawBooks_df['Rating'].astype(str).str.extract(r'(\d+\.?\d*)')  # Extracting numeric part from Rating
rawBooks_df['Num Of Reviews'] = rawBooks_df['Num Of Reviews'].astype(str).str.extract(r'(\d+)')  # Extracting numeric part from Num Of Reviews

rawBooks_df['Price'] = rawBooks_df['Price'].astype(float)

# Display the cleaned DataFrame
rawBooks_df
```
