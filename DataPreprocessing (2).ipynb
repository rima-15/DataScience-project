# Data Preprocessing

### - Handling Missing Values


```python
import pandas as pd
import re

rawBooks_df = pd.read_csv("raw_books.CSV")

# pattren for n/a 
pattern = '\u200en\u200e/\u200ea\u200e'

# Search for this pattren
specific_hidden_char_rows = rawBooks_df[rawBooks_df.apply(lambda x: x.astype(str).str.contains(pattern, na=False)).any(axis=1)]

# display
specific_hidden_char_rows
```

```python

# replace this pattren with 'Unknown'
for col in rawBooks_df.columns:
    rawBooks_df[col] = rawBooks_df[col].apply(
        lambda x: 'Unknown' if isinstance(x, str) and pattern in x else x
    )
```

```python
# Check
unkown_author = rawBooks_df[rawBooks_df.apply(lambda x: x.astype(str).str.contains('Unknown', na=False)).any(axis=1)]
unkown_author
```

```python
missing_rows = rawBooks_df[(rawBooks_df == 'Null').any(axis=1) | 
                        (rawBooks_df == 'Unknown').any(axis=1) | 
                        (rawBooks_df.isna()).any(axis=1)]

#  Count the number of "Null", "Unknown", and NaN values in each column
missing_counts = ((rawBooks_df == 'Null') | (rawBooks_df == 'Unknown')).sum() + rawBooks_df.isna().sum()

#  Display the count of missing values in each column
print("\nCount of missing values ('Null', 'Unknown', or NaN) in each column:")
missing_counts
```

```python
print("Rows with 'Null', 'Unknown', or empty (NaN) values:")
missing_rows
```

```python
rawBooks_df['Rating'] = rawBooks_df['Rating'].replace('Null', float('nan')).fillna(0)
rawBooks_df['Num Of Reviews'] = rawBooks_df['Num Of Reviews'].replace('Null', float('nan')).fillna(0)
rawBooks_df = rawBooks_df.dropna(subset=['Author'])

rawBooks_df[['Rating', 'Num Of Reviews','Author']] 

# NOTE: Need to rearrrange the index after the drop 
# NOTE : Need to explain the 3 cases of Missing value 
# NOTE : Need a code of pattren n/a ?
```

### - Format for Price , Rating , Num Of Reviews 

```python
# Cleaning the columns
rawBooks_df['Price'] = rawBooks_df['Price'].astype(str).str.extract(r'(\d+\.?\d*)')  # Extracting numeric part from Price
rawBooks_df['Rating'] = rawBooks_df['Rating'].astype(str).str.extract(r'(\d+\.?\d*)')  # Extracting numeric part from Rating
rawBooks_df['Num Of Reviews'] = rawBooks_df['Num Of Reviews'].astype(str).str.extract(r'(\d+)')  # Extracting numeric part from Num Of Reviews

rawBooks_df['Price'] = rawBooks_df['Price'].astype(float)

# Display the cleaned DataFrame
rawBooks_df
```

### - Normalization
for Num Of Rebiewa


```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 5))

# Normalize the "Price" and "Num of Reviews" columns
rawBooks_df[['Num Of Reviews']] = scaler.fit_transform(rawBooks_df[['Num Of Reviews']])

```

```python
# Display the cleaned DataFrame
rawBooks_df
```

### - Discretization 
for Price 


```python

# Get the min and max of the Price column to understand the range
min_price = rawBooks_df['Price'].min()
max_price = rawBooks_df['Price'].max()

print(f"Min Price: {min_price}, Max Price: {max_price}")

# Define bin edges using quantiles (5 bins based on price distribution)
bin_edges = [min_price,
             rawBooks_df['Price'].quantile(0.2),
             rawBooks_df['Price'].quantile(0.4),
             rawBooks_df['Price'].quantile(0.6),
             rawBooks_df['Price'].quantile(0.8),
             max_price]

# Define labels for each bin
bin_labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']

# Apply discretization using pd.cut()
rawBooks_df['Price'] = pd.cut(rawBooks_df['Price'], bins=bin_edges, labels=bin_labels, right=False)



rawBooks_df

```

### - Encoding 
for Price 


```python
# Encode labels to integers
label_encoding = {'Very Low': 0, 'Low': 1, 'Medium': 2, 'High': 3, 'Very High': 4}
rawBooks_df['Price'] = rawBooks_df['Price'].map(label_encoding)
rawBooks_df
```

### Genre Cleaning and Standardization


```python
# Step 1: Clean the Genre column
# Remove duplicate commas and trim whitespace
rawBooks_df['Genre'] = rawBooks_df['Genre'].str.replace(',,', ',').str.strip()

# Remove unnecessary genres from both the beginning and the end of the genre string
unnecessary_genres = ['Best Sellers', 'Jarir Publications Books', 'New Arrivals', 'Bargain Books', 'E-Books', 'Audio & Video CD/DVD']
for unwanted in unnecessary_genres:
    rawBooks_df['Genre'] = rawBooks_df['Genre'].str.replace(f'{unwanted}, ', '', regex=False)
    rawBooks_df['Genre'] = rawBooks_df['Genre'].str.replace(f', {unwanted}', '', regex=False)
    rawBooks_df['Genre'] = rawBooks_df['Genre'].str.replace(unwanted, '', regex=False).str.replace(',,', ',').str.strip(', ')

# Step 2: Define the mapping for genre standardization
genre_mapping = {
    # Fiction Genres
    'Fiction & Literature': 'Fiction Genres',
    'Literary Fiction': 'Fiction Genres',
    'Historical Fiction': 'Fiction Genres',
    'Psychological Fiction': 'Fiction Genres',
    'Genre Fiction': 'Fiction Genres',
    'Mysteries': 'Fiction Genres',
    'Mysteries & Thrillers for Young Adults': 'Fiction Genres',
    'Mysteries & Detective Stories for Children': 'Fiction Genres',
    'Science Fiction': 'Fiction Genres',
    'Fantasy for Young Adults': 'Fiction Genres',
    'Romantic Fantasy': 'Fiction Genres',
    'Historical Mystery': 'Fiction Genres',
    'Psychological Thrillers': 'Fiction Genres',
    'Comics & Graphic Novels': 'Fiction Genres',
    'Manga Comics & Graphic Novels': 'Fiction Genres',
    'Humor': 'Fiction Genres',
    'Humorous Fiction': 'Fiction Genres',

    # Non-Fiction Genres (mapped under specific categories)
    'Self-Help': 'Non-Fiction Genres',
    'Self Development': 'Non-Fiction Genres',
    'Self-Esteem': 'Non-Fiction Genres',
    'Motivational Self-Help': 'Non-Fiction Genres',
    'Self-Help for Anger Management': 'Non-Fiction Genres',
    'Health': 'Non-Fiction Genres',
    'Fitness & Dieting': 'Non-Fiction Genres',
    'Diets': 'Non-Fiction Genres',
    'Diseases & Physical Ailments': 'Non-Fiction Genres',
    'Psychological Therapy': 'Non-Fiction Genres',
    'Business & Management': 'Non-Fiction Genres',
    'Business Management': 'Non-Fiction Genres',
    'Finance': 'Non-Fiction Genres',
    'Politics & Social Sciences': 'Non-Fiction Genres',
    'Forensic Psychology': 'Non-Fiction Genres',
    'Psychology & Counseling': 'Non-Fiction Genres',
    'Philosophy': 'Non-Fiction Genres',
    'Education Studies & Teaching': 'Non-Fiction Genres',
    'Graduate School Test Guides': 'Non-Fiction Genres',
    'School Textbooks': 'Non-Fiction Genres',
    'Higher & Continuing Education': 'Non-Fiction Genres',
    'Foreign Language Study & Reference': 'Non-Fiction Genres',
    'Textbooks and Study Guides': 'Non-Fiction Genres',
    'Technology': 'Non-Fiction Genres',
    'Digital Art': 'Non-Fiction Genres',

    # Children & Young Adults
    'Children\'s Early Learning Books on the Alphabet': 'Children & Young Adults',
    'Early Learning for Children': 'Children & Young Adults',
    'Explore the World Books for Children': 'Children & Young Adults',
    'Children Books': 'Children & Young Adults',
    'Education Studies & Teaching': 'Children & Young Adults',
    'Children\'s Books on Country & Farm Life': 'Children & Young Adults',
    'Children\'s Books on Family Life': 'Children & Young Adults',
    'Children\'s Books on Social Skills': 'Children & Young Adults',
    'Children\'s Books on Crafts & Hobbies': 'Children & Young Adults',
    'Literature & Fiction for Children': 'Children & Young Adults',
    'Fantasy for Young Adults': 'Children & Young Adults',
    'Mysteries & Thrillers for Young Adults': 'Children & Young Adults',

    # Arts, Crafts, & Hobbies
    'Drawing': 'Arts, Crafts, & Hobbies',
    'Digital Art': 'Arts, Crafts, & Hobbies',
    'Crafts': 'Arts, Crafts, & Hobbies',
    'Crafts & Hobbies': 'Arts, Crafts, & Hobbies',
    'Hobbies & Home': 'Arts, Crafts, & Hobbies',
    'Sports & Outdoors': 'Arts, Crafts, & Hobbies',
    'Baking': 'Arts, Crafts, & Hobbies',

    # Religion & Spirituality
    'Islamic & Religion Books': 'Religion & Spirituality',
    'Islam': 'Religion & Spirituality',

    # History & Culture
    'History': 'History & Culture',
    'History & Geography': 'History & Culture',
    'Antiques & Collectibles': 'History & Culture',

    # Poetry & Essays
    'Poetry & Literary Essays': 'Poetry & Essays',
    'Poetry Anthologies': 'Poetry & Essays',
    'Literary Essays & Correspondence': 'Poetry & Essays',

    # Parenting & Family
    'Family & Relationships': 'Parenting & Family',
    'Parenting & Family': 'Parenting & Family',

    # Cookbooks & Food
    'Cooking & Culinary Arts': 'Cookbooks & Food',
    'Celebrity & TV Show Cookbooks': 'Cookbooks & Food',
    'Baking': 'Cookbooks & Food',
}

# Step 3: Define a function to clean and map genres
def clean_genre_list(genre_str):
    if pd.isna(genre_str):  # Check if the value is NaN or None
        return None
    # Split the genres, remove unnecessary genres, and strip whitespace
    genres = [g.strip() for g in genre_str.split(',') if g.strip() and g not in unnecessary_genres]
    # Map genres to standardized values based on the genre_mapping dictionary
    mapped_genres = set(genre_mapping.get(g, g) for g in genres)
    # If multiple valid genres exist, return the primary category
    main_categories_priority = ['Fiction Genres', 'Non-Fiction Genres', 'Children & Young Adults', 'Arts, Crafts, & Hobbies',
                                'Religion & Spirituality', 'History & Culture', 'Poetry & Essays', 'Parenting & Family', 'Cookbooks & Food']
    for category in main_categories_priority:
        if category in mapped_genres:
            return category
    return ', '.join(mapped_genres) if mapped_genres else None

# Step 4: Apply the cleaning and mapping function to each row
rawBooks_df['Genre'] = rawBooks_df['Genre'].apply(clean_genre_list)

# Replace rows with no valid genres with empty strings
rawBooks_df['Genre'] = rawBooks_df['Genre'].fillna('')

# Display the cleaned DataFrame
rawBooks_df
```
